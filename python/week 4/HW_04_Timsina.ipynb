### Problem 1: Sample Mean and Variance (look back)

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 16, 26, 12, 17, 22, 14, 12, 21, 16])
n=len(x)
mean = np.sum(x)/n
diff = x- mean
diff_sqrd = diff ** 2
var = (sum(diff_sqrd)/n-1)
print("Sample Mean:\t",mean)
print("Sample Variance:",var)
```

```python
mean_np = np.mean(x)
var_np = np.var(x,ddof=1)
print("Sample Mean:\t",mean_np)
print("Sample Variance:",var_np)
```

### Problem 2: Scoring a Regression Model

```python
def find_sse(true_y,pred_y):
    n=len(true_y)
    return(sum((true_y - pred_y)**2))
    
```

### Problem 4: Classification Report

```python
def  classification_report(true_y,pred_y):
    classes = np.unique(true_y)
    accuracy = find_accuracy(true_y,pred_y)
    TP = np.sum(pred_y == classes[1])
    FP = np.sum(true_y == classes[0])
    TN = np.sum(pred_y==classes[1])
    FN = np.sum(true_y==classes[0])
    positive_precision = TP/(TP+FP)
    positive_recall = TP/(TP+FN)
    negative_precision = TN/(TN+FN)
    negative_recall = TN/(TN+FP)
    print("Positive Class :\t ", classes[1])
    print("Negetive Class :\t ", classes[0])
    print("")
    print("Accuracy:\t\t ",accuracy)
    print("Positive Precision :\t ", positive_precision)
    print("Positive Recall :\t ", positive_recall)
    print("Negetive Precision :\t ", negative_precision)
    print("Negative Recall :\t ", negative_recall)
    
```

### Problem 6: Stochastic Linear Relationships

```python
np.random.seed(1)
x_vals = (np.random.normal(10,2, size=200))
errors = (np.random.normal(0,1.2, size=200))
y_vals = 5.1 + 0.9 * x_vals + errors

plt.figure(figsize=[8,6])
plt.scatter(x=x_vals, y=y_vals, s=60, alpha=0.8, 
            color='red', edgecolor='black')
plt.xlabel('X_values')
plt.ylabel('Y values')
plt.title('relationship between x values and Y values')
plt.show()
```

```python
diff_x = x_vals-np.mean(x_vals)
diff_y = y_vals-np.mean(y_vals)
r = np.sum((diff_x)*(diff_y))/(np.sum((diff_x)**2) * np.sum((diff_y)**2) )**0.5

print("Correlation between X and Y:",r)
```

### Problem 7: Relationship between Life Expectancy and Per Capita GDP


```python
 import pandas as pd
 df = pd.read_csv('gapminder_data.txt', sep='\t')
 country = df.country.values
 year = df.year.values
 continent = df.continent.values
 population = df.population.values
 life_exp = df.life_exp.values
 pcgdp = df.gdp_per_cap.values
 gini = df.gini.values
 df = None
```

```python
continent_list = ['africa', 'americas', 'asia','europe']
color_list = ['cornflowerblue', 'aquamarine', 'papayawhip','silver']
plt.figure(figsize=[8,6])

temp_pcgdp = []
temp_life_exp= []

for i in range(0, len(continent_list)):
    sel = ((continent_list[i] == continent) & (year == 2018))
    current_continent = continent_list[i]
    for j in range(0,len(sel)):
        if(sel[j]==True):
            temp_pcgdp.append(pcgdp[j])
            temp_life_exp.append(life_exp[j])
    #print(temp_pcgdp)
    #print(temp_life_exp)
    plt.scatter(x=np.log(temp_pcgdp),label=continent_list[i].title(), y=temp_life_exp, s=100,alpha=0.7,color=color_list[i], edgecolor='black')
    plt.xlim([6, 12])
    plt.ylim([45, 90])
    plt.xlabel('Natural Log of Per Capita GDP')
    plt.ylabel('Life Expectancy')
    plt.title('Life Expectency vs Per Capita GDP (2018)')
    temp_pcgdp = []
    temp_life_exp= []
plt.legend()
plt.show()
```
